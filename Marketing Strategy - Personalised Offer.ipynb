```python
missing_columns =['no_visited_Cold drinks','Restaur_spend_less_than20','no_visited_bars','car','no_Take-aways','Restaur_spend_greater_than20']
for i in missing_columns:
    print(train_data[i].value_counts())
```

```python
# convert ordinal categorical data into numerical data
train_data['no_visited_Cold drinks'].replace({'never':-1,'less1':0,'1~3':2,'4~8':6,'gt8':9},inplace=True)
train_data['Restaur_spend_less_than20'].replace({'never':-1,'less1':0,'1~3':2,'4~8':6,'gt8':9},inplace=True)
train_data['no_visited_bars'].replace({'never':-1,'less1':0,'1~3':2,'4~8':6,'gt8':9},inplace=True)
train_data['no_Take-aways'].replace({'never':-1,'less1':0,'1~3':2,'4~8':6,'gt8':9},inplace=True)
train_data['Restaur_spend_greater_than20'].replace({'never':-1,'less1':0,'1~3':2,'4~8':6,'gt8':9},inplace=True)
```

```python
# Dropping 'Job/job industry' column because more number of unique values and 'car' column because more missing values
train_data.drop(columns=['Job/Job Industry','car'],inplace=True)
print(train_data.columns)
print(train_data.shape)
```

```python
train_data['offer expiration'].replace({'2days':48,'10hours':10},inplace=True)
```

```python
# ordinal data
train_data['income_range'].replace({'Less than ₹12500':0,'₹12500 - ₹24999':1,'₹25000 - ₹37499':2,
                                       '₹37500 - ₹49999':3,'₹50000 - ₹62499':4,'₹62500 - ₹74999':5,
                                       '₹75000 - ₹87499':6,'₹87500 - ₹99999':7,'₹100000 or More':8},inplace=True)
train_data['age'].replace({'below21':20,'21':21,'26':26,'31':31,'36':36,'41':41,'46':46,
                          '50plus':51},inplace=True)
train_data['Qualification'].replace({'Some High School':0,'High School Graduate':1,'Some college - no degree':2,
                'Bachelors degree':3,'Associates degree':4,'Graduate degree (Masters or Doctorate)':5},inplace=True)
```

```python
# Handling nominal data
import numpy as np
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
```

```python
ohe.fit_transform(train_data[['Marital Status','restaurant type','gender','Customer type','Climate','drop location']]).toarray()
```

```python

feature_array =ohe.fit_transform(train_data[['Marital Status','restaurant type','gender','Customer type','Climate','drop location']]).toarray()
feature_labels =ohe.categories_
feature_labels
```

```python
new_features=[]
for i in range(len(feature_labels)):
    for j in range(len(feature_labels[i])):
        new_features.append(feature_labels[i][j])
print(new_features)
```

```python
nominal_data=pd.DataFrame(feature_array,columns=new_features)
```

```python
final_train=pd.concat([train_data,nominal_data],axis=1)
final_train.head()
```

```python
final_train.drop(columns=['Marital Status','restaurant type','gender','Customer type','Climate','drop location'],inplace=True)
```

```python
final_train['Offer Accepted'].replace({'Yes':1,'No':0},inplace=True)
final_train.shape
```

```python
# train_data['Maribackend_managerstal Status'] ---- 5 nominal variables
# train_data['restaurant type'].----- 5 nominal values
#train_data['gender'] ------ 2 values
# train_data['Customer type'] ---- 4 values
# train_data['Climate'] ---------- 3 values
# train_data['drop location'] --- 3 values
```

```python
# Label
final_train['Offer Accepted']
```

```python
# Handling missing values
final1_train = final_train.drop(columns='Offer Accepted')
print(final1_train.shape)
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp1 = imp.fit(final1_train)
final2_train=imp1.transform(final1_train)
final2_train
```

```python
final2_train
```

```python
x = final2_train
y = final_train['Offer Accepted']
x.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```
